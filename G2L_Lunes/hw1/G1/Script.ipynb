```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el documento

imagenes_url = '/content/sample_data/california_housing_test.csv'
vl = pd.read_csv(imagenes_url)

# variables para examinar

variables = ["housing_median_age", "total_rooms", "total_bedrooms", "population",
             "households", "median_income", "median_house_value"]

# configuracion de clase

sns.set_style("whitegrid")

# Crear figuras con 3 graficos por variables

brief, core = plt.subplots(nrows=7, ncols=3, figsize=(18, 28))
core = core.flatten()


for i, var in enumerate(variables):
     # Histograma
    sns.histplot(vl[var], bins=30, kde=True, ax=core[i * 3])
    core[i * 3].set_title(f"Histograma de {var}")
    core[i * 3].set_xlabel(var)
    core[i * 3].set_ylabel("Frecuencia")


    # Boxplot (Medidas de centralidad)
    sns.boxplot(x=vl[var], ax=core[i * 3 + 1])
    core[i * 3 + 1].set_title(f"Boxplot de {var}")
    core[i * 3 + 1].set_xlabel(var)

   # Violinplot (Medidas de dispersin)
    sns.violinplot(x=vl[var], ax=core[i * 3 + 2])
    core[i * 3 + 2].set_title(f"Violinplot de {var}")
    core[i * 3 + 2].set_xlabel(var)

# Ajustar el bosquejo
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

vl = pd.read_csv('/content/sample_data/california_housing_test.csv')


# Configuracion General de Graficos

plt.figure(figsize=(18, 6))

# Grfico 1: Edad media de la vivienda contra nmero total de habitaciones

plt.subplot(1, 3, 1)
sns.scatterplot(x=vl['housing_median_age'], y=vl['total_rooms'])
plt.title('housing_median_age vs total_rooms')
plt.xlabel('housing_median_age')
plt.ylabel('total_rooms')

# Grfico 2: Edad media de la vivienda contra valor medio de la casa

plt.subplot(1, 3, 2)
sns.scatterplot(x=vl['housing_median_age'], y=vl['median_house_value'])
plt.title('housing_median_age vs median_house_value')
plt.xlabel('housing_median_age')
plt.ylabel('median_house_value')


# Grfico 3: Nmero total de habitaciones contra valor medio de la casa

plt.subplot(1, 3, 3)
sns.scatterplot(x=vl['total_rooms'], y=vl['median_house_value'])
plt.title('total_rooms vs median_house_value')
plt.xlabel('total_rooms')
plt.ylabel('median_house_value')

plt.tight_layout()
plt.show()

# Calcular la matriz de valoracin

calculate_matrix = vl[['housing_median_age', 'total_rooms', 'median_house_value']].corr()

# Visualizar la matriz de valoracin con un mapa de calor

plt.figure(figsize=(6, 5))
sns.heatmap(calculate_matrix, annot=True, cmap='coolwarm', fmt =".2f",linewidths=0.5)
plt.title('matrix de calcuacin entre las variables')
plt.show()
```
